### matplotlib practise

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label="quadratic")
plt.plot(x, x**3, label = "cubic")

plt.title("plot")
plt.xlabel("x number")
plt.ylabel("y number")

plt.legend()
#plt.show()
```
